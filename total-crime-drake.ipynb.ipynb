# Crime and Police Activities in Different Neighborhoods

```python
import pandas as pd
import numpy as np
import geopandas
%matplotlib inline
import matplotlib.pyplot as plt
```

For this data set the unit used will be called TC for "Total Crime." The less TC a neighborhood has the better the neighborhood is. TC will be calculated from how many crimes occur in certain neighboorhoods. For the purposes of finding the best neighborhood the top 5 neighborhoods will be inspected for how many crimes have been committed.

```python
newCrime = pd.read_csv("2016-recent.csv", index_col="PK", low_memory = False)
oldCrime = pd.read_csv("archive-police-blotter.csv", index_col="PK", low_memory = False)
newCrime["INCIDENTNEIGHBORHOOD"].unique
```

```python
len(newCrime["INCIDENTNEIGHBORHOOD"].unique())
```

```python
crimeCounts = newCrime['INCIDENTNEIGHBORHOOD'].value_counts().sort_values(ascending=True)
focusedCrime = crimeCounts.head(13)
crimeCounts.head(13)
```

The top 10 contenders are show here because Outside County, Outside State, and Swisshelm Park are not actually neighborhoods so it is redundant for our purposes. This means out of the 99 neighborhoods within this Pittsburgh dataset the ones with the least crimes are Mt. Oliver Boro, Mt. Olviver Neighborhood, Chartiers City, Troy Hiller-Herrs Island, Ridgemont, New Homestead, East Carnegie, Arlington Heights, Mount Olver, and Summer Hill.  The next graphs plot a helpful graph to visualize the TC of each neighborhood.

```python
oldCrimeCounts = oldCrime['INCIDENTNEIGHBORHOOD'].value_counts().sort_values(ascending=True)
oldCrimeCounts.head(10)
```

```python
g1 = crimeCounts.head(13).plot(kind = "bar")
plt.xticks(rotation=90, horizontalalignment = "center")
plt.xlabel("Neighborhoods")
plt.ylabel("Total Crime")
plt.title("Total Number of Crimes From 2016-Present")
```

```python
oldCrimeCounts.head(10).plot(kind = "bar")
plt.xticks(rotation=90, horizontalalignment = "center")
plt.xlabel("Neighborhoods")
plt.ylabel("Total Crime")
plt.title("Total Number of Crimes from 2009 - 2016")
```

```python
crimeCounts.tail(10).plot(kind = "bar")
```

From these graphs you can see that the difference in the numbers of crimes are astronomical. When considering the best Neighborhood i would want to live in one with as little crime as possible.

These graphs represent both the more recent crimes and the older crimes within the neighborhoods of Pittsburgh. It also shows the most crime ridden neighborhoods so that one can see the difference between the most versus the least.
